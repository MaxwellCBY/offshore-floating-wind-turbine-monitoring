```python
import pynmea2
import random

def generate_nmea_sentence():
    # Generate random GPS coordinates and speed
    latitude = random.uniform(-90.0, 90.0)
    longitude = random.uniform(-180.0, 180.0)
    speed = random.uniform(0.0, 10.0)

    # Create a GPVTG sentence with the speed information
    vtg_sentence = pynmea2.GPVTG(
        "T",
        "M",
        "0.00",
        "N",
        f"{speed:.2f}",
        "K",
        "N"
    )

    # Create a GPGGA sentence with the GPS location information
    gga_sentence = pynmea2.GPGGA(
        "000000.00",
        f"{latitude:.6f}",
        "N",
        f"{longitude:.6f}",
        "W",
        "1",
        "08",
        "1.0",
        "M",
        "-33.0",
        "M",
        "",
        ""
    )

    # Concatenate the two sentences into a single NMEA sentence
    nmea_sentence = f"{vtg_sentence}\r\n{gga_sentence}\r\n"

    return nmea_sentence

# Write 10 randomly generated NMEA sentences to a file
with open("gps_data.txt", "w") as f:
    for i in range(10):
        sentence = generate_nmea_sentence()
        f.write(sentence)
```

```python
import pynmea2
import datetime

# create a GPVTG message with some example data
message = pynmea2.GPVTG('054.7', 'T', '034.4', 'M', '005.5', 'N', '010.2', 'K')

# get the current time
current_time = datetime.datetime.now()

# create a NMEA sentence with the GPVTG message and the current time
nmea_sentence = pynmea2.NMEACommand('GP', 'VTG', (current_time.strftime('%H%M%S.%f'), message))

# print the NMEA sentence
print(str(nmea_sentence))
```
